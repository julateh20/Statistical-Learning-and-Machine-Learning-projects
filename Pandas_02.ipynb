## Query

- allow using an expression
- same queries for multiple dataframe
- faster way of doing the selection than the rest of the method 
- query return the value directly


```python
df = pd.DataFrame({'A': range(1, 6),
...                    'B': range(10, 0, -2),
...                    'C C': range(10, 5, -1)})
```

```python
df
```

```python
df.query('A>B')
```

```python
eval("2+3")
```

```python
c=3
df.query("A>@c") # use the "@" symbol for accessing the variable
```

```python
df.query('B == `C C`')  #---- so the column with spaces should be written inside ` ` 
```

```python
df[df.A==df.B]
```

```python
df.query('A>1 & B>5')
```
